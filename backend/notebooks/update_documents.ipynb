## Connect to Meilisearch

```python
import meilisearch

meilisearch_client = meilisearch.Client("http://localhost:7700", "aStrongMasterKey")
print("Successfully connected to Meilisearch!")
```

## Count documents

Let's retrieve every document that has the blog name `ElasticsearchCollapseSearchResults`.

```python
index = meilisearch_client.index("articles")
blog_name = "ElasticsearchCollapseSearchResults"
response = index.get_documents({"filter": f"name = '{blog_name}'", "limit": 100})
chunks = response.results
len(chunks)
```

Before updating the documents, let's review the current view counts.

```python
for chunk in chunks:
    view_count = chunk.view_count
    print(f"Chunk ID: {chunk.id}, View Count: {view_count}")
```

All chunks must have the same view count. Let's update the view counts for all chunks.

```python
new_view_count = chunks[0].view_count + 1
documents_to_update = [
    {"id": chunk.id, "view_count": new_view_count} for chunk in chunks
]
documents_to_update
```

Now, let's update the documents in MeiliSearch.

```python
task = index.update_documents(documents_to_update)
meilisearch_client.wait_for_task(task.task_uid)
```

Let's verify that the documents have been updated.

```python
response = index.get_documents({"filter": f"name = '{blog_name}'", "limit": 100})
chunks = response.results

for chunk in chunks:
    view_count = chunk.view_count
    print(f"Chunk ID: {chunk.id}, View Count: {view_count}")
```
