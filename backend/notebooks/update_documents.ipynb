## Connect to Meilisearch

```python
import meilisearch

meilisearch_client = meilisearch.Client("http://localhost:7700", "aStrongMasterKey")
print("Successfully connected to Meilisearch!")
```

## Count documents

Let's retrieve the document that has the blog name `elasticsearch-collapse-search-results`.

```python
index = meilisearch_client.index("articles")
blog_name = "elasticsearch-collapse-search-results"
response = index.get_documents({"filter": f"name = '{blog_name}'"})
results = response.results
if results:
    document = results[0]
    print(f"Document found: {document}")
else:
    document = None
```

Before updating the documents, let's review the current view counts.

```python
if document is not None:
    view_count = document.view_count
    print(f"Document ID: {document.id}, View Count: {view_count}")
```

All chunks must have the same view count. Let's update the view counts for all chunks.

```python
if document is not None:
    new_view_count = document.view_count + 1
    documents_to_update = [{"id": document.id, "view_count": new_view_count}]
```

Now, let's update the documents in MeiliSearch.

```python
task = index.update_documents(documents_to_update)
meilisearch_client.wait_for_task(task.task_uid)
```

Let's verify that the documents have been updated.

```python
response = index.get_documents({"filter": f"name = '{blog_name}'"})
results = response.results

if results:
    updated_document = results[0]
    print(f"Updated view count: {updated_document.view_count}")
```
