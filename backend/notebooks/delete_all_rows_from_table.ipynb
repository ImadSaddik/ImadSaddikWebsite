## Delete all Rows

```python
import sqlite3

connection = sqlite3.connect("../visitors.db")
with connection:
    cursor = connection.cursor()
    cursor.execute("DELETE FROM visitors;")
    connection.commit()
    print("All rows from 'visitors' table have been deleted.")
```

## Verify the rows count

```python
import sqlite3

connection = sqlite3.connect("../visitors.db")
with connection:
    cursor = connection.cursor()
    cursor.execute("SELECT COUNT(*) FROM visitors;")
    count = cursor.fetchone()[0]
    print(f"Number of rows in 'visitors' table: {count}")
```
