## Connect to Meilisearch

```python
import meilisearch

meilisearch_client = meilisearch.Client("http://localhost:7700", "aStrongMasterKey")
print("Successfully connected to Meilisearch!")
```

## Create an index

```python
index_name = "articles"
index = meilisearch_client.index(index_name)
print(f"Index '{index_name}' is ready.")
```

## Index documents

### Create documents

```python
import uuid

from datetime import datetime

uuid_value = uuid.uuid4()
creation_date_unix = int(datetime.strptime("27/12/2025", "%d/%m/%Y").timestamp())

article_name = "LocalAIStackOnLinux"
file_path = f"../../frontend/src/blogs/{article_name}/content.md"
with open(file_path, "r") as file:
    file_content = file.read()

documents = [
    {
        "id": str(uuid_value),
        "name": article_name,
        "tags": ["Linux", "AI", "LLM", "llama.cpp", "LibreChat", "Local AI", "llama-swap"],
        "creation_date": creation_date_unix,
        "year": "2025",
        "title": "How to build your own local AI stack on Linux with llama.cpp, llama-swap, LibreChat and more",
        "type": "blog-post",
        "content": file_content,
        "view_count": 0,
        "read_count": 0,
        "claps_count": 0,
    },
]
```

```python
import json

with open("documents.json", "w") as json_file:
    json.dump(documents, json_file, indent=4)
```

### Index

```python
task = index.add_documents(documents)

print(f"Adding documents... Task UID: {task.task_uid}")
meilisearch_client.wait_for_task(task.task_uid)
print("Documents have been added successfully!")
```
