## Connect to Meilisearch

```python
import meilisearch

meilisearch_client = meilisearch.Client("http://localhost:7700", "aStrongMasterKey")
print("Successfully connected to Meilisearch!")
```

## Connect to Meilisearch

```python
index = meilisearch_client.index("articles")
stats = index.get_stats()
document_count = stats.number_of_documents

print(f"The index '{index.uid}' contains {document_count} documents.")
```

## Add new field to all documents

Since I don't have a lot of documents, I will fetch all documents at once. For larger datasets, consider using pagination.

```python
default_value = 0
new_field_name = "read_count"

response = index.get_documents({"limit": 1000, "offset": 0})
documents = response.results

documents_to_update = [
    {"id": document.id, "read_count": default_value} for document in documents
]

task = index.update_documents(documents_to_update)
meilisearch_client.wait_for_task(task.task_uid)
print(f"Successfully added '{new_field_name}' to all documents!")
```

## Fetch documents again and verify the new field

```python
response = index.get_documents({"limit": 1000, "offset": 0})
documents = response.results

for document in documents:
    keys = document.__dict__.keys()
    if new_field_name not in keys:
        print(f"Document ID {document.id} is missing the '{new_field_name}' field.")
```
