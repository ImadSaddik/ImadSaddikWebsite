## Connect to Meilisearch

```python
import meilisearch

meilisearch_client = meilisearch.Client("http://localhost:7700", "aStrongMasterKey")
print("Successfully connected to Meilisearch!")
```

## Configure the index

Setting the sortable attributes makes Meilisearch sort the search results based on those attributes.

```python
index_name = "test"
index = meilisearch_client.index(index_name)
task = index.update_sortable_attributes(
    [
        "creation_date",
        "view_count",
    ]
)
meilisearch_client.wait_for_task(task.task_uid)
print("Sortable attributes have been updated successfully!")
```

## Search


### No query

```python
from datetime import datetime

search_results = index.search("", {"sort": ["view_count:asc"]})
for hit in search_results["hits"]:
    date_str = datetime.fromtimestamp(hit["creation_date"]).strftime("%Y-%m-%d")
    print(f"- {hit['title']} (Date: {date_str}) - {hit['view_count']} views")
```

```python
search_results = index.search("", {"sort": ["view_count:desc"]})
for hit in search_results["hits"]:
    date_str = datetime.fromtimestamp(hit["creation_date"]).strftime("%Y-%m-%d")
    print(f"- {hit['title']} (Date: {date_str}) - {hit['view_count']} views")
```
